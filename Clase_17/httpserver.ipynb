## Protocolo HTTP
https://datatracker.ietf.org/doc/html/rfc2616

El Protocolo de Transferencia de Hipertexto (HTTP), en inglés "Hypertext Transfer Protocol," es un protocolo de comunicación utilizado en la World Wide Web (WWW) para la transferencia de datos entre un cliente y un servidor. Fue desarrollado para permitir la recuperación de recursos, como páginas web, imágenes, videos y otros tipos de archivos, desde servidores web y su presentación en el navegador del cliente.

HTTP opera en un modelo de solicitud-respuesta, donde un cliente, generalmente un navegador web, realiza solicitudes a un servidor web para obtener recursos específicos. El servidor procesa estas solicitudes y responde con los recursos solicitados, junto con información adicional, como metadatos y códigos de estado HTTP que indican el resultado de la solicitud.

El protocolo HTTP se basa en una arquitectura cliente-servidor, donde el cliente inicia la comunicación enviando una solicitud HTTP al servidor, y el servidor responde con la información solicitada. Cada solicitud y respuesta HTTP consta de un conjunto de encabezados (headers) que transportan metadatos sobre la solicitud o la respuesta, seguidos opcionalmente de un cuerpo que puede contener datos, como el contenido de una página web.

HTTP es un protocolo sin estado, lo que significa que cada solicitud se procesa de manera independiente, sin conocimiento del estado anterior del cliente o del servidor. Para mantener el estado en aplicaciones web, se utilizan mecanismos adicionales como cookies y sesiones.

Además del protocolo HTTP básico, existen varias versiones, siendo las más comunes HTTP/1.1 y HTTP/2. Estas versiones han introducido mejoras en la eficiencia de la transferencia de datos y en la velocidad de carga de las páginas web.

## Los métodos HTTP 
Los métodos HTTP, también conocidos como verbos HTTP, son acciones que indican la intención del cliente web o del usuario cuando realiza una solicitud a un servidor web. Cada solicitud HTTP debe incluir un método HTTP para especificar qué acción se debe realizar en el recurso solicitado. Los métodos HTTP son parte fundamental del protocolo HTTP y ayudan a definir el comportamiento de la solicitud. Los métodos HTTP más comunes son:


| **Método HTTP** | **Descripción**                                                                              |
|:-----------------|:--------------------------------------------------------------------------------------------------------|
| GET             | El método GET se utiliza para recuperar información del servidor. Cuando un cliente envía una solicitud GET, está pidiendo al servidor que le envíe un recurso específico, como una página web, una imagen o un archivo. Este método no realiza cambios en el servidor ni en los datos.                                    |
| POST            | El método POST se utiliza para enviar datos al servidor para ser procesados. Por ejemplo, cuando un usuario llena un formulario en línea y hace clic en "Enviar", los datos del formulario se envían al servidor utilizando POST. Este método se utiliza para crear nuevos recursos en el servidor o actualizar datos existentes y puede contener información confidencial.
|
| PUT             | El método PUT se utiliza para actualizar un recurso existente en el servidor o para crear uno nuevo si no existe. Es idempotente, lo que significa que realizar la misma solicitud varias veces no debería tener efectos secundarios diferentes.
|
| DELETE          | El método DELETE se utiliza para eliminar un recurso específico en el servidor. Al igual que PUT, es idempotente, lo que significa que eliminar el mismo recurso varias veces no debería causar problemas.    |
| HEAD            | El método HEAD es similar a GET, pero solicita solo los encabezados de respuesta y no el cuerpo del recurso. Se utiliza para obtener información sobre el recurso, como el tamaño o la fecha de modificación, sin descargar todo el contenido.
|
| PATCH           | El método PATCH se utiliza para realizar modificaciones parciales en un recurso. En lugar de reemplazar completamente el recurso, el servidor aplica los cambios proporcionados en la solicitud PATCH. Esto es útil cuando se necesita actualizar solo una parte de un recurso.
|
| OPTIONS         | El método OPTIONS se utiliza para solicitar información sobre las opciones de comunicación disponibles para un recurso en particular. Esto incluye los métodos HTTP permitidos, los encabezados compatibles y otras capacidades del servidor.
|
| CONNECT         | El método CONNECT se utiliza para establecer una conexión de red con un servidor web a través de un proxy. Se utiliza en casos de túneles de comunicación segura, como conexiones SSL/TLS a través de un proxy.
|
| TRACE           | El método TRACE se utiliza para realizar una prueba de bucle de retroalimentación en la ruta hacia el servidor. No se usa comúnmente y a menudo está deshabilitado por razones de seguridad.
|

## Ejemplos de solicitudes GET y POST

#### Explicación de los encabezados:

- **GET /pagina.html HTTP/1.1**: Indica que estamos haciendo una solicitud GET para obtener el recurso "pagina.html" utilizando HTTP versión 1.1.
- **Host: www.ejemplo.com**: Especifica el nombre de dominio del servidor al que se realiza la solicitud.
- **User-Agent**: Proporciona información sobre el navegador o el agente de usuario que realiza la solicitud.
- **Accept**: Indica los tipos de medios que el cliente puede procesar. En este caso, el cliente acepta HTML, XHTML y XML.
- **Accept-Language**: Especifica los idiomas que el cliente prefiere para la respuesta.
- **Accept-Encoding**: Indica los métodos de compresión que el cliente admite.
- **Connection**: keep-alive: Solicita al servidor que mantenga abierta la conexión para posibles solicitudes adicionales.

El servidor procesará esta solicitud y responderá con el contenido de "pagina.html" si está disponible. Esta es una solicitud GET típica en HTTP.

### Códigos de estado HTTP
Los códigos de estado HTTP se utilizan para comunicar el resultado de una solicitud HTTP entre el servidor y el cliente, lo que permite a los clientes comprender si la solicitud se completó correctamente o si ocurrió un error.

#### 2xx - Respuestas Exitosas
Estos códigos indican que la solicitud fue recibida, entendida y aceptada con éxito.

**200 - OK**: La solicitud se ha procesado correctamente y se ha devuelto una respuesta.

**201 - Creado**: La solicitud se ha procesado con éxito y ha resultado en la creación de un nuevo recurso.

**204 - Sin contenido**: La solicitud se procesó con éxito, pero no se devolvió contenido en la respuesta.

#### 3xx - Redirecciones
Estos códigos indican que el cliente debe tomar alguna acción adicional para completar la solicitud.

**301 - Movido permanentemente**: La ubicación de la solicitud ha cambiado permanentemente, y el cliente debe actualizar su URL.

**302 - Encontrado (Found)**: La solicitud se redirige temporalmente a una nueva ubicación.

#### 4xx - Errores del Cliente
Estos códigos indican que ha habido un error por parte del cliente en la solicitud.

**400 - Solicitud incorrecta**: La solicitud del cliente es incorrecta o no se puede entender.

**401 - No autorizado**: El cliente no está autorizado para acceder al recurso solicitado.

**403 - Prohibido**: El cliente no tiene permisos para acceder al recurso solicitado.

**404 - No encontrado**: El recurso solicitado no se ha encontrado en el servidor.

#### 5xx - Errores del Servidor
Estos códigos indican que ha habido un error por parte del servidor al procesar la solicitud del cliente.

**500 - Error interno del servidor**: Indica que el servidor ha encontrado un error interno al procesar la solicitud y no puede completarla.

**503 - Servicio no disponible**: El servidor no puede manejar la solicitud en este momento debido a sobrecarga o mantenimiento.

## El módulo http.server
El módulo http.server de Python es una parte de la biblioteca estándar de Python que proporciona una forma sencilla de crear servidores web HTTP en Python. Es una herramienta muy útil para el desarrollo web, pruebas locales o para compartir archivos a través de una red. Este módulo permite a los desarrolladores crear rápidamente un servidor web local sin la necesidad de configuraciones complejas.

Aquí hay una breve introducción a algunas de las características clave del módulo http.server:

1. Servidor Web Simple: El módulo http.server permite crear un servidor web HTTP básico en Python. Puede ser utilizado para servir archivos estáticos, como páginas HTML, imágenes, archivos CSS o JavaScript, desde tu sistema de archivos local.

2. Fácil de Usar: Configurar un servidor web local con este módulo es tan simple como ejecutar un comando en la línea de comandos. No se necesita escribir un código personalizado para iniciar un servidor web básico.

3. Soporte para Diferentes Versiones de HTTP: El módulo http.server admite tanto HTTP/1.0 como HTTP/1.1, lo que significa que es posible interactuar con él utilizando navegadores web modernos o herramientas que utilizan estas versiones de HTTP.

4. Puerto Personalizable: Se puede especificar el puerto en el que se desea ejecutar el servidor web. De forma predeterminada, se ejecuta en el puerto 8000, pero es posible cambiarlo fácilmente.

5. Listado de Directorios: Si, al acceder a una URL de un directorio que no contiene un archivo de índice (como index.html), el servidor web generará automáticamente un listado de archivos y directorios en ese directorio para que los usuarios puedan navegar.

6. Soporte para CGI (Common Gateway Interface): Es posible ejecutar scripts CGI en tu servidor web con el módulo http.server.

7. Ampliamente Personalizable: Aunque es simple de usar, el módulo http.server también es personalizable. Pueden extenderse sus clases y modificar su comportamiento según tus necesidades.

8. Seguridad: Es importante tener en cuenta que el módulo http.server no es adecuado para entornos de producción y no debe utilizarse para exponer servidores web públicos. Está destinado principalmente a entornos de desarrollo y pruebas locales.

## Programando un servidor http (chiquito)

````python
import http.server
import socketserver

PORT = 8000

Handler = http.server.SimpleHTTPRequestHandler

with socketserver.TCPServer(("", PORT), Handler) as httpd:
    print("serving at port", PORT)
    httpd.serve_forever()
````

## Programando un servidor http (más grandecito)
https://docs.python.org/3/library/http.server.html

````python
import http.server
import socketserver

# GET / HTTP/1.1

PORT = 1111


class handler_manual (http.server.BaseHTTPRequestHandler):
    def do_GET(self):
        print("REQUEST: ", self.requestline)
        self.send_response(200)
        self.send_header("Content-Type", "text/html")
        self.end_headers()
        self.wfile.write(b'hola mundo GET\n')

    def do_POST(self):
        print("REQUEST: ", self.requestline)
        self.send_response(200)
        self.send_header("Content-Type", "text/html")
        self.end_headers()
        self.wfile.write(b'hola mundo POST\n')

socketserver.TCPServer.allow_reuse_address = True

#myhttphandler = http.server.BaseHTTPRequestHandler
#myhttphandler = http.server.SimpleHTTPRequestHandler
myhttphandler = handler_manual


#httpd = socketserver.TCPServer(("", PORT), myhttphandler)
httpd = http.server.HTTPServer(("", PORT), myhttphandler)
#httpd = http.server.ThreadingHTTPServer(("", PORT), myhttphandler)

print(f"Opening httpd server at port {PORT}")

httpd.serve_forever()

httpd.shutdown()
````

##### Uso de Apache Benchmark (ab)

Apache Benchmark (ab) es una herramienta de línea de comandos utilizada para realizar pruebas de carga en servidores web. Forma parte del proyecto Apache y resulta útil para medir el rendimiento del servidor web bajo condiciones de carga.

###### Instalación

1. **En Linux**: Apache Benchmark generalmente se incluye en el paquete `apache2-utils`. Se puede instalar ejecutando:

   ```bash
   sudo apt-get install apache2-utils
   ```

2. **En macOS**: Se puede instalar Apache Benchmark utilizando Homebrew con el siguiente comando:

   ```bash
   brew install httpd
   ```

   Este comando instalará `ab` junto con otros componentes de Apache.

3. **En Windows**: Apache Benchmark se incluye con la distribución de Apache HTTP Server. Descargando e instalando Apache desde el sitio oficial, se encontrará el ejecutable `ab.exe` en el directorio `bin`.

###### Uso Básico

El comando básico para utilizar Apache Benchmark es:

```bash
ab -n <número_de_solicitudes> -c <concurrencia> <URL>
```

- `-n <número_de_solicitudes>`: Especifica el número total de solicitudes a realizar.
- `-c <concurrencia>`: Especifica cuántas solicitudes se enviarán de forma simultánea.
- `<URL>`: La URL del servidor web que se desea probar.

###### Ejemplos de Uso

1. **Prueba básica**: Para realizar 100 solicitudes a un servidor web con una concurrencia de 10:

   ```bash
   ab -n 100 -c 10 http://www.ejemplo.com/
   ```

2. **Prueba con autenticación básica**: Si el servidor requiere autenticación básica, se puede usar la opción `-A`:

   ```bash
   ab -n 100 -c 10 -A usuario:contraseña http://www.ejemplo.com/
   ```

3. **Prueba enviando datos con POST**: Para enviar datos a través de POST, se utiliza la opción `-p` para especificar un archivo con los datos y `-T` para indicar el tipo de contenido:

   ```bash
   ab -n 100 -c 10 -p datos.txt -T application/x-www-form-urlencoded http://www.ejemplo.com/
   ```

   Donde `datos.txt` contiene los datos a enviar, por ejemplo, `param1=valor1&param2=valor2`.

4. **Prueba con encabezados personalizados**: Para agregar encabezados HTTP personalizados, se utiliza la opción `-H`:

   ```bash
   ab -n 100 -c 10 -H "Authorization: Bearer token_aqui" http://www.ejemplo.com/
   ```

###### Interpretación de Resultados

Después de ejecutar una prueba, Apache Benchmark proporciona varias métricas útiles:

- **Requests per second**: Cantidad de solicitudes que el servidor pudo manejar por segundo.
- **Time per request**: Tiempo promedio que el servidor tomó para manejar una solicitud individual.
- **Transfer rate**: Cantidad de datos transferidos por segundo.
- **Connection Times**: Tiempos detallados sobre la conexión, desde la conexión inicial hasta la recepción de datos completos.

Estas métricas ayudan a evaluar la capacidad de manejo de carga del servidor web y a identificar posibles cuellos de botella.

###### Consejos para Realizar Pruebas

1. **Realizar múltiples pruebas** para obtener una media representativa.
2. **Aumentar gradualmente la carga** para identificar el punto en que el rendimiento comienza a degradarse.
3. **Probar desde múltiples ubicaciones** si es posible, para simular cargas de usuarios distribuidos geográficamente.
4. **Monitorizar los recursos del servidor** (CPU, RAM, I/O) durante las pruebas para entender el impacto.



# Uvicorn: Servidor Web ASGI para Producción

## Introducción

En el ámbito del desarrollo web, la elección del servidor es crucial para el rendimiento y la escalabilidad de las aplicaciones. Mientras que servidores como el módulo `http.server` de Python son útiles para desarrollo y pruebas locales, no son adecuados para entornos de producción. En este contexto, Uvicorn emerge como una solución robusta y eficiente para el despliegue de aplicaciones web en producción.

## ¿Qué es Uvicorn?

Uvicorn es un servidor web ASGI (Asynchronous Server Gateway Interface) de alto rendimiento. Está diseñado específicamente para ser utilizado en entornos de producción, ofreciendo una serie de ventajas sobre los servidores web tradicionales.

## Características principales

1. **Alto rendimiento**: Uvicorn está construido sobre la librería `uvloop`, que proporciona un bucle de eventos extremadamente rápido. Este diseño permite a Uvicorn superar en rendimiento a muchos servidores web sincrónicos tradicionales.

2. **Soporte para ASGI**: Implementa el estándar ASGI (Asynchronous Server Gateway Interface), lo que permite el manejo de solicitudes de manera asíncrona. Esta característica resulta en una mejor utilización de los recursos del sistema y una mayor capacidad para manejar conexiones concurrentes.

3. **Escalabilidad**: Gracias a su arquitectura asíncrona, Uvicorn puede manejar un gran número de conexiones simultáneas. Esta capacidad lo hace especialmente adecuado para aplicaciones web de alto tráfico o que requieren procesamiento en tiempo real.

4. **Compatibilidad**: Se integra perfectamente con frameworks web modernos como FastAPI, Starlette y Django (cuando se configura para ASGI). Esta compatibilidad facilita la transición de aplicaciones existentes a un entorno de alto rendimiento.

5. **Configuración flexible**: Ofrece una amplia gama de opciones para ajustar el rendimiento según las necesidades específicas de cada aplicación. Esto incluye configuraciones para el número de trabajadores, el manejo de reinicio automático, y más.

6. **Soporte para WebSockets**: Además de HTTP, Uvicorn proporciona soporte nativo para WebSockets, permitiendo la implementación de aplicaciones en tiempo real de manera eficiente.

## Instalación y uso básico

Para comenzar a utilizar Uvicorn, primero es necesario instalarlo. Esto se puede hacer fácilmente utilizando pip:

```bash
pip install uvicorn
```

Un ejemplo básico de cómo iniciar una aplicación ASGI con Uvicorn sería:

```python
import uvicorn

async def app(scope, receive, send):
    assert scope['type'] == 'http'
    await send({
        'type': 'http.response.start',
        'status': 200,
        'headers': [
            [b'content-type', b'text/plain'],
        ],
    })
    await send({
        'type': 'http.response.body',
        'body': b'Hello, World!',
    })

if __name__ == "__main__":
    uvicorn.run("nombre_del_modulo:app", host="127.0.0.1", port=8000, reload=True)
```

En este ejemplo, se define una aplicación ASGI simple que responde con "Hello, World!" a todas las solicitudes. Luego, se utiliza Uvicorn para ejecutar esta aplicación en el puerto 8000.

## Integración con frameworks web

Uvicorn brilla especialmente cuando se utiliza en conjunto con frameworks web modernos. Por ejemplo, con FastAPI, una de las combinaciones más populares, se puede crear una API de alto rendimiento de la siguiente manera:

```python
from fastapi import FastAPI
import uvicorn

app = FastAPI()

@app.get("/")
async def root():
    return {"message": "Hello World"}

if __name__ == "__main__":
    uvicorn.run(app, host="0.0.0.0", port=8000)
```

## Configuraciones avanzadas

Uvicorn ofrece una variedad de opciones de configuración para optimizar el rendimiento:

1. **Trabajadores**: Se puede especificar el número de procesos de trabajo para manejar las solicitudes.
   ```bash
   uvicorn app:app --workers 4
   ```

2. **Recarga automática**: Útil durante el desarrollo para recargar automáticamente la aplicación cuando se detectan cambios en el código.
   ```bash
   uvicorn app:app --reload
   ```

3. **SSL/TLS**: Para habilitar HTTPS directamente en Uvicorn.
   ```bash
   uvicorn app:app --ssl-keyfile key.pem --ssl-certfile cert.pem
   ```

4. **Logging**: Configuración de diferentes niveles de logging para depuración y monitoreo.
   ```bash
   uvicorn app:app --log-level debug
   ```

## Consideraciones de seguridad y producción

Aunque Uvicorn está diseñado para entornos de producción, en un escenario de despliegue real, generalmente se utiliza detrás de un servidor proxy inverso como Nginx. Esta configuración proporciona beneficios adicionales:

1. **Equilibrio de carga**: Distribuye el tráfico entre múltiples instancias de la aplicación.
2. **Terminación SSL**: Maneja la encriptación SSL/TLS, descargando esta tarea de Uvicorn.
3. **Caching**: Proporciona caching a nivel de servidor web para mejorar el rendimiento.
4. **Seguridad adicional**: Ofrece una capa adicional de protección contra ataques comunes.


## Modelo de concurrencia

Uvicorn ofrece flexibilidad en su modelo de concurrencia, permitiendo el uso tanto de multihilado como de multiproceso, dependiendo de las necesidades específicas de la aplicación y del entorno de despliegue.

### Multiproceso

Por defecto, cuando se configura Uvicorn para utilizar múltiples workers, utiliza un modelo de multiproceso:

1. **Workers**: Cada worker es un proceso separado. Esto se configura con el parámetro `--workers`:

   ```bash
   uvicorn app:app --workers 4
   ```

   En este caso, Uvicorn lanzará 4 procesos separados, cada uno manejando conexiones de forma independiente.

2. **Ventajas del multiproceso**:
   - Mejor utilización de múltiples núcleos de CPU.
   - Aislamiento entre procesos, lo que puede mejorar la estabilidad.
   - Capacidad para manejar cargas de trabajo intensivas en CPU de manera más eficiente.

3. **Consideraciones**:
   - Mayor uso de memoria, ya que cada proceso tiene su propio espacio de memoria.
   - No comparten estado directamente entre procesos.

### Multihilado

Aunque Uvicorn es principalmente asíncrono y no utiliza múltiples hilos para manejar solicitudes concurrentes dentro de un solo proceso, puede beneficiarse del multihilado en ciertos escenarios:

1. **Threadpool**: Uvicorn utiliza un threadpool para operaciones que podrían bloquear el bucle de eventos, como operaciones de E/S sincrónicas.

2. **Integración con bibliotecas sincrónicas**: Cuando se trabaja con código o bibliotecas que no son asíncronas, el threadpool permite que estas operaciones no bloqueen todo el servidor.

3. **Configuración del threadpool**: Se puede ajustar el tamaño del threadpool utilizando la opción `--workers`:

   ```bash
   uvicorn app:app --workers 1 --threads 8
   ```

   Esto crea un proceso con un pool de 8 hilos para manejar operaciones potencialmente bloqueantes.

### Modelo híbrido

En entornos de producción, es común utilizar un modelo híbrido que combina multiproceso y multihilado:

1. **Múltiples procesos**: Se lanzan varios procesos worker para aprovechar múltiples núcleos de CPU.

2. **Threadpool por proceso**: Cada proceso tiene su propio threadpool para manejar operaciones bloqueantes.

3. **Ejemplo de configuración**:

   ```bash
   uvicorn app:app --workers 4 --threads 8
   ```

   Esto crearía 4 procesos, cada uno con un pool de 8 hilos.

### Elección del modelo

La elección entre multihilado, multiproceso o un modelo híbrido depende de varios factores:

- **Naturaleza de la aplicación**: Aplicaciones con operaciones intensivas en CPU se benefician más del multiproceso.
- **Recursos del servidor**: Servidores con múltiples núcleos pueden aprovechar mejor el multiproceso.
- **Tipo de carga**: Para cargas con muchas operaciones de E/S, un modelo híbrido puede ser más eficiente.
- **Requisitos de memoria**: El multiproceso consume más memoria, lo que debe considerarse en entornos con recursos limitados.

En última instancia, la configuración óptima a menudo se determina mediante pruebas de rendimiento y ajustes basados en las características específicas de la aplicación y el entorno de despliegue.
